# 프로그램 설치

* miniconda 다운로드 및 설치 (WIndows용 64bit)
* https://conda.io/miniconda.html


* All User용으로 설치



* 설치가 완료되면 [시작] 버튼 클릭
* Anaconda Prompt 실행


* 아래 명령어 입력하여 ml 이라는 환경을 만듬
* conda create -n ml python=3.6 


* 아래와 같이 입력하여 방금 만든 콘다 환경을 활성화 시킨다.
* 콘다 환경을 활성화 시키면 모든 파이썬의 라이브러리(=패키지)들이 ml이라는 환경에 설치가 된다.
* activate ml


* conda install 명령어로 필요한 라이브러리를 설치한다.
* conda install pandas matplotlib seaborn tensorflow scikit-learn jupyter


* conda install ipykernel 
* 설치를 완료한 뒤


* ipykernel jupyter notebook 커널스팩 설치
* python -m ipykernel install --name ml --display-name "Python 3(ml)" 


* 완료되면 jupyter notebook 실행
* jupyter notebook
